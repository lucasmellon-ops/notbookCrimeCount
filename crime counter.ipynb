```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\Lucas\Desktop\Crime_Data_from_2020_to_Present.csv"
data = pd.read_csv(file_path)

```

```python
data['Date Rptd'] = pd.to_datetime(data['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')
data['month'] = data['Date Rptd'].dt.month
data['year_month'] = data['Date Rptd'].dt.to_period('M')

```

```python
print(data.columns)
```

```python

monthly_crimes = data.groupby('month').size().reset_index(name='count')

```

```python
plt.plot(monthly_crimes['month'], monthly_crimes['count'])
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Crimes Reported by Month')
plt.show()

```
